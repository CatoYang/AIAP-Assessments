#### Domain Knowledge: CO2 measurement methods

The naming of features suggests different sensor types for CO2 detection. <br>

1. `CO2_InfraredSensor` is likely to be using NDIR (Non-Dispersive **Infrared**) sensors <br>
The Principle of detection is infrared absorption spectroscopy. <br>
CO2 absorbs IR light at a specific wavelength, A IR source shines through a measuring chamber into a calibrated IR detector. <br>
The differential is proportional to the Co2 concentration<br>

This sensor measures CO2 Specifically and is robust over time <br>
Sensitive to direct light and IR rich illumination, possibly influenced by `Ambient Light Level`<br>

2. `CO2_ElectroChemicalSensor` is as its name suggests uses a **Electrochemical** sensor <br>
The Principle of detection is a Redox chemical reaction generating current. <br>
The sensor contains electrodes submerged in an electrolyte. <br>
When a gas (e.g., CO) enters the sensor, it reacts at the working electrode <br>
This flow of electrons is measured as an electric current, proportional to the gas concentration. <br>

This sensor is less reliable/accurate for CO2 as CO2 is chemically stable and less prone to redox reactions <br>
There is a risk of drift and cross-sensitivity due to other chemicals <br>
The method is a indirect measurement of CO2 (chemical mix can be tuned to CO2)<br>

3. `MetalOxideSensor_Unitn` as its name suggests uses a MOS (**Metal Oxide Semiconductor**) sensor<br>
The Principle of detection is Chemoresistive measurement.<br>
The sensor uses a heated metal oxide surface<br>
In clean air, oxygen molecules absorb onto the surface and steal electrons, creating a high resistance layer<br>
When reducing gases (e.g., VOCs, CO, H₂) are present, they react with the absorbed oxygen, releasing the electrons back and lowering resistance.<br>
The Measured current is used as a variable along with humidity and temperature to give a inferred reading<br>

As CO2 does not strongly react with MOS surfaces its not directly detected and is cross-sensitive to other gasses<br>
The sensor also requires calibration<br>
The method is a inferred measurement of CO2 <br>

##### What does this mean?
It tells us that `CO2_InfraredSensor` is the most reliable sensor as compared to the other two. It measures CO2 **directly** where as the other two merely measures it **indirectly** or **infers** it. As `CO2_ElectroChemicalSensor` and `MetalOxideSensor_Unitn` use in some form or another a chemical reaction its prone to deterioration over time and we have no means of factoring that into analysis.<br>

Most sensor units are typically packaged together into a single device. To limit the scope of the analysis we assume this to be true. <br>
There is additional possibility if we are told there is a positional data we would have to revisit this EDA with sensor positioning <br>

`MetalOxideSensor_Unit` could possibly be measuring something other than CO2 considering we have not been told what they are measuring, they can include any form of reducing gases depending on the sensor material.

We also do not know the unit of measurement, the common unit is ppm (parts per million) and the range is consistent with `CO2_ElectroChemicalSensor` and `MetalOxideSensor_Unitn`.<br>

The Sensor type for `CO_GasSensor` is also undetermined, but we can probably tell if there is a strong relationship between the feature and `Ambient Light Level` for NDIR sensors and `Temperature` & `Humidity` for Electro chemical or Metal Oxide

#### Data Correction in `Temperature`

The list of attributes listed temperature's unit of measurement to be in celsius <br>
But the range has major outliers, the maximum range is 298.93 which should not be possible <br>

```python
# Examining Temperature
sns.histplot(df_data['Temperature'])
```

**visual comparison** of the shape of the histogram its likely that the outlier is caused by data input measured in kelvin instead of celcius.
Its highly unlikely its measured in fahrenheit most ambient temperatures would inhabit the 0-100 range.
We can transform the data and re-examine

```python
# Seperating data and transforming
df_data['IsKelvin'] = df_data['Temperature'] > 150

df_data['Temperature_Corrected'] = df_data.apply(
    lambda row: row['Temperature'] - 273.15 if row['IsKelvin'] else row['Temperature'],
    axis=1
)
```

```python
# Plotting new subset against bigger cluster
sns.kdeplot(df_data['Temperature'][~df_data['IsKelvin']], label='Original Celsius')
sns.kdeplot(df_data['Temperature_Corrected'][df_data['IsKelvin']], label='Corrected Kelvin → Celsius')
plt.legend()
plt.title('Comparison of Celsius vs Corrected Kelvin Data')
plt.xlabel('Temperature (°C)')
plt.show()
```

```python
# Creating descriptive statistic
print("Original Celsius Values:")
print(df_data.loc[~df_data['IsKelvin'], 'Temperature'].describe())

print("\nCorrected Kelvin Values:")
print(df_data.loc[df_data['IsKelvin'], 'Temperature_Corrected'].describe())
```

```python
# Defining variable
celsius = df_data.loc[~df_data['IsKelvin'], 'Temperature']
kelvin_corrected = df_data.loc[df_data['IsKelvin'], 'Temperature_Corrected']

# t-test : Checks if average values differ based on normal distrobution, the shape slightly follows one
t_stat_T, p_ttest_T = ttest_ind(celsius, kelvin_corrected, equal_var=False)
print(f"T-statistic: {t_stat_T:.3f}, p-value: {p_ttest_T:.3f}")

# KS test : Checks for overall behavour on full distrobutions
ks_stat_T, p_kstest_T = ks_2samp(celsius, kelvin_corrected)
print(f"KS statistic: {ks_stat_T:.3f}, p-value: {p_kstest_T:.3f}")
```

**Analysis** <br>
Look at the **descriptive statistics** the mean and range are very close to one another<br>
The percentile spread also line up closely. <br>
Student's T-Test and Kolmogoro-Smirnov Test (KS test) values who the means are nearly identical. The p-values over 0.05 which signifies that there is no significant difference<br>
Based on **statistical comparison** we can presume its safe to merge the data.

```python
# Overwrite Temperature with corrected values where IsKelvin is True
df_data.loc[df_data['IsKelvin'], 'Temperature'] = df_data.loc[df_data['IsKelvin'], 'Temperature_Corrected']

# drop the helper columns
df_data.drop(columns=['IsKelvin', 'Temperature_Corrected'], inplace=True)

# Checking if changes are acccurate
df_data['Temperature'].agg(['min', 'max'])
```

##### Initial Heuristical Approach

We might be able to discern the type of error quickly examining the shape of the outliers.

```python
# Examining the possibility of the outlier's shape matching
# Humidity values below 0
neg_humidity = df_data[df_data['Humidity'] < 0]

# Humidity values between 100 and 150 (inclusive)
lower_set = df_data[(df_data['Humidity'] > 100) & (df_data['Humidity'] <= 150)]

# Humidity values above 150
upper_set = df_data[df_data['Humidity'] > 150]

#  Humidity above 100
total_set = df_data[(df_data['Humidity'] > 100)]
```

Splitting the data to see if the negative values match the shape of the other
I suspect that if one half matches the shape we can better conclude they belong to the same error/data subset

```python
# Treating data
neg_humidity_abs = neg_humidity.copy()
neg_humidity_abs['Humidity'] = neg_humidity_abs['Humidity'].abs()

neg_humidity_shift50 = neg_humidity.copy()
neg_humidity_shift50['Humidity'] = neg_humidity_shift50['Humidity'] + 50

# Subtract 100 from 'Humidity' values
lower_set = lower_set.copy()
lower_set['Humidity'] = lower_set['Humidity'] - 100

upper_set = upper_set.copy()
upper_set['Humidity'] = upper_set['Humidity'] - 100

total_set = total_set.copy()
total_set['Humidity'] = total_set['Humidity'] - 100
```

```python
plt.figure(figsize=(12, 6))

# Plot each distribution separately
sns.kdeplot(neg_humidity_abs['Humidity'], label='Negative (Abs)', fill=True, alpha=0.5)
sns.kdeplot(neg_humidity_shift50['Humidity'], label='Negative (Shift50)', fill=True, alpha=0.5)
sns.kdeplot(lower_set['Humidity'], label='100–150 → shifted', fill=True, alpha=0.5)
sns.kdeplot(upper_set['Humidity'], label='>150 → shifted', fill=True, alpha=0.5)
sns.kdeplot(total_set['Humidity'], label='full range', fill=True, alpha=0.5)

plt.title('Comparison of Adjusted Humidity Distributions')
plt.xlabel('Humidity')
plt.ylabel('Density')
plt.legend()
plt.tight_layout()
plt.show()
```

There is a slight mirror shape wise of
- negative values shifted 50 right
- values above 150

we can test it with statistical comparison after transformings

```python
# We use Abs dataframe as its already transformed
neg_humidity_abs_shift50 = neg_humidity_abs.copy()
neg_humidity_abs_shift50['Humidity'] = neg_humidity_abs_shift50['Humidity'] +50

# Comparison of neg_humidity_abs_shift50  and upper_Set
t_stat_RH, p_val_RH = ttest_ind(
    neg_humidity_abs_shift50['Humidity'],
    upper_set['Humidity'],
    equal_var=False
)
ks_stat_RH, ks_p_RH = ks_2samp(
    neg_humidity_abs_shift50['Humidity'],
    upper_set['Humidity']
)
print(f"T-statistic: {t_stat_RH:.3f}, p-value: {p_val_RH:.3f}")
print(f"KS statistic: {ks_stat_RH:.3f}, p-value: {ks_p_RH:.3f}")
```

P values are above 0.05, we can conclude that the distributions are not statistically significant and they likely come from the same underlying distribution. Im inclined to just apply a shift. <br>

#### Relabeling and Ordinal Mapping for `Ambient Light Level `'s NaN values

```python
# Evaluating statistical significance of excluding values
print(f"Missing values percentage in Ambient Light Level : {df_data['Ambient Light Level'].isna().sum()/100}%")
```

This is a significant portion of our data and heavy considerations should be done before considering to drop.

```python
# Fill NaNs in CO_GasSensor with missing
df_plot = df_data.copy()
df_plot['Ambient Light Level'] = df_plot['Ambient Light Level'].fillna('Missing')

# Get all categorical columns except 'CO_GasSensor'
cat_cols = df_plot.select_dtypes(include=['object', 'category']).columns.tolist()

for col in cat_cols:
    # Fill NaNs for the other categorical column
    df_plot[col] = df_plot[col].fillna('Missing')

    plt.figure(figsize=(12, 6))
    sns.countplot(data=df_plot, x='Ambient Light Level', hue=col)
    plt.xticks(rotation=45)
    plt.title(f'Ambient Light Level count by {col}')
    plt.tight_layout()
    plt.show()
```

At a glance the missing values's distribution is very consistent across categorical categories.

```python
# Chi Square test
for col in cat_cols:
    contingency = pd.crosstab(df_plot['Ambient Light Level'], df_plot[col])
    chi2, p, dof, expected = chi2_contingency(contingency)

    print(f'Chi-square test for Ambient Light Level vs {col}')
    print(f'p-value: {p:.4f}')
    if p < 0.05:
        print("Likely dependent (statistically significant)")
    else:
        print("Likely independent (not statistically significant)")
    print()
```

Based on Chi squared test, 'missing' is significant and cannot be dropped. <br>
There is a clear linearity to 'missing' and it can be place consistently between bright and very bright.<br>
As there the relationship between light and most other features should be linear, heuristically its safe to order it so. <br>
The statistical significance to `CO_GasSensor` can be explained if we consider the possibility that **the sensor is a NDIR type and the light's lumination frequency coincides with the absorption frequency of CO.**

```python
# Relabeling
df_data['Ambient Light Level'] = df_data['Ambient Light Level'].fillna('brighter')

ordinal_map = {
    'verydim': 'verydim',
    'dim': 'dim',
    'moderate': 'moderate',
    'bright': 'bright',
    'brighter': 'brighter',  # relabeled
    'verybright': 'brightest'  # pushed one level up
}

# Apply mapping
df_data['Ambient Light Level'] = df_data['Ambient Light Level'].replace(ordinal_map)

# Define the order of categories explicitly (including new 'brightest' level)
ordered_levels = ['verydim', 'dim', 'moderate', 'bright', 'brighter', 'brightest']

# Convert to ordered categorical dtype
df_data['Ambient Light Level'] = pd.Categorical(
    df_data['Ambient Light Level'], 
    categories=ordered_levels, 
    ordered=True
)
```

At a glance — ignoring outliers — the spread of humidity appears unusually linear and narrow <br>
This shows that temperature does not have a noticable effect on humidity.
This is unexpected. From a conventional perspective — assuming all else equal — the relationship should be exponential <br>

Using the distribution of the arden buck equation to see how it should look like

```python
# Finding out the Temperature range
print(df_data['Temperature'].min(), df_data['Temperature'].max())
```

```python
# Temperature range (°C)
T = np.linspace(0, 100, 500)

# Assume constant actual vapor pressure (e = 10 hPa) This is a good middle ground for ambient air at 50% RH and 20 C
e_actual = 10  # in hPa

# Arden Buck equation for saturation vapor pressure
e_s = 6.1121 * np.exp((17.502 * T) / (T + 240.97))

# Relative Humidity as a function of temperature
RH = (e_actual / e_s) * 100

# Plot
plt.figure(figsize=(10, 6))
plt.plot(T, RH, label='RH vs Temp (Arden Buck)', color='blue')

# Comfort zone and selected temperature range
plt.axhspan(40, 60, color='lightgreen', alpha=0.3, label='Comfort RH Range (40–60%)')
plt.axvline(x=14.94, color='red', linestyle='--', label='T = 14.94°C')
plt.axvline(x=25.99, color='orange', linestyle='--', label='T = 25.99°C')

plt.xlabel('Temperature (°C)')
plt.ylabel('Relative Humidity (%)')
plt.title('Relative Humidity vs Temperature\nUsing Arden Buck Equation (e = 10 hPa)')
plt.ylim(0, 110)
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

From the arden buck equation we can see that relative humidity should rapidly change with temperature.<br>
This reveals to us that HVAC is working to adjust the water vapour pressure and thus the humidity to compensate for any fluctuations in temperature and humidity. <br>

During rain the ground level ambient air conditions should be saturated generally at or near 100%. We **should expect** (heuristics) a few data points' humidity to spike beyond the comfort range, especially during modes when HVAC is `ventilationonly`  and perhaps during `off` or `maintenance mode`. But it clearly does not show, there would be a heavy tailed distribution towards 100% during rain. <br>
Even if the environments is isolated, any occupants should passively give off water vapour during respiration unless hooked to a ventilator.<br>

From this we can make a few observations/assertations
- We can ignore the possibility of rain affecting the data
- The Environment is relatively isolated from the outdoors.
- HVAC is working actively to keep the room within the comfort RH range.
- Humidity's outliers is not caused by HVAC settings

We can also conclude that its unlikely that the outliers obvserved in humidity belong to the full 0 - 100 range. 
We could either drop the values in outliers or re-correct them to the cluster's range.

But to enable scaling in the future retraining, it would be best to scale from 0 - 100 regardless.

```python
# Mi for our target feature 'Activity Level' byu class
df_encoded = df_data.copy()
for col in df_encoded.select_dtypes(include='category').columns:
    df_encoded[col] = df_encoded[col].cat.codes

X = df_encoded.drop(columns=['Activity Level'])
y = df_encoded['Activity Level']

classes = y.unique()
mi_per_class = {}

for cls in classes:
    # Create a binary target for this class vs rest
    y_binary = (y == cls).astype(int)
    
    mi_scores = mutual_info_classif(X, y_binary, discrete_features='auto', random_state=42)
    mi_per_class[cls] = pd.Series(mi_scores, index=X.columns).sort_values(ascending=False)

# Display MI scores per class
for cls, mi_scores in mi_per_class.items():
    print(f"\nMutual Information Scores for class {cls}:")
    print(mi_scores.head(10))
```

```python
# Models
from sklearn.ensemble import RandomForestClassifier #,HistGradientBoostingClassifier,
#from sklearn.neural_network import MLPClassifier
#from sklearn.linear_model import LogisticRegression
#from sklearn.svm import SVC
from catboost import CatBoostClassifier
from xgboost import XGBClassifier

# Tools
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.utils.class_weight import compute_class_weight
from sklearn.model_selection import train_test_split , RandomizedSearchCV
from collections import defaultdict
from sklearn.metrics import (
    accuracy_score,
    f1_score,
    fbeta_score,
    precision_score,
    recall_score,
    log_loss,
    confusion_matrix,
    classification_report,
)
```

Our target variable is `activity level`, a categorical feature with 3 unique values which equates to a multiclass (3) classification problem

I have opted to use 3 tree-based ensemble models <br>

1. Random Forest Classifier (bagged/bootstrap aggregate) <br>
   robust and needs less tuning
2. XGBoost Classifier (gradient boosing)<br>
   robust, Build in regularization
3. CatBoost (Ordered Boosting) <br>
   Same as gradient boosting, but with better handling of categorical features

The use of tree-based models is intentional, because we do not know the proper ranges for the numerical data it can adversely affect the results of most models. Tree based models are robust when handling this. <br>
Other Reasoning for the absence of other models
- Neural Networks : Requires more hyperparameter tuning and data just to outperform simpler models. Also suffers from lack of numerical ranges
- Support Vector Machines : Not scalable in the future as datasets increase
- K-Nearest Neighbour : Potentially good, but can suffer since we dont have proper ranges.

```python
# Define models individually
xgb_clf = XGBClassifier(
    objective='binary:logistic',
    n_estimators=1000,
    max_depth=7,
    learning_rate=0.05,
    subsample=0.8,
    colsample_bytree=0.8, 
    gamma=1,
    reg_alpha=0.1,
    reg_lambda=1.0,
    scale_pos_weight=1,  
    use_label_encoder=False,
    eval_metric='logloss',
    random_state=13,
    verbosity=0
)

cat_clf = CatBoostClassifier(
    iterations=1000,
    depth=7,
    learning_rate=0.05,
    l2_leaf_reg=3.0,
    border_count=64,
    subsample=0.8,
    bootstrap_type='Bernoulli',
    loss_function='MultiClass',
    eval_metric='Accuracy',
    custom_metric=['F1'],
    random_seed=13,
    verbose=0,
    class_weights=[2, 1, 2]
)

rf_clf = RandomForestClassifier(
    criterion="gini",
    n_estimators=1750,
    max_depth=7,
    min_samples_split=6,
    min_samples_leaf=6,
    max_features='sqrt',
    oob_score=True,
    random_state=13,
    class_weight={0: 2, 1: 1, 2: 2},
    verbose=0
)

# Now store them in a list
models = [xgb_clf, cat_clf, rf_clf]
```

```python
def run_models(dataframe):
    target_col = "Activity Level"

    # Split data
    train, test = train_test_split(dataframe, test_size=0.3, random_state=42, shuffle=True)

    # Separate features and target
    X_train = train.drop(columns=[target_col])
    y_train = train[target_col]
    X_test = test.drop(columns=[target_col])
    y_test = test[target_col]

    # Identify categorical columns in features
    categorical_cols = X_train.select_dtypes(include=['category']).columns.tolist()

    # Setup OneHotEncoder for categorical features
    preprocessor = ColumnTransformer(
        transformers=[
            ('cat', OneHotEncoder(handle_unknown='ignore', sparse_output=False), categorical_cols)
        ],
        remainder='passthrough'
    )


    # Encode features
    X_train_enc = preprocessor.fit_transform(X_train)
    X_test_enc = preprocessor.transform(X_test)

    # Encode target labels
    y_train_enc = y_train.cat.codes.values
    y_test_enc = y_test.cat.codes.values

    results = defaultdict(dict)

    # Train and evaluate models
    for model in models:
        model.fit(X_train_enc, y_train_enc)
        y_pred = model.predict(X_test_enc)

        model_name = type(model).__name__
        results[model_name]["Accuracy"] = f"{accuracy_score(y_test_enc, y_pred)*100:.2f}%"
        results[model_name]["F1 (macro)"] = f"{f1_score(y_test_enc, y_pred, average='macro'):.2f}"
        results[model_name]["F1 (weighted)"] = f"{f1_score(y_test_enc, y_pred, average='weighted'):.2f}"
        results[model_name]["F2 Score"] = f"{fbeta_score(y_test_enc, y_pred, beta=2, average='macro'):.2f}"
        results[model_name]["Precision (macro)"] = f"{precision_score(y_test_enc, y_pred,zero_division=0, average='macro'):.2f}"
        results[model_name]["Recall (macro)"] = f"{recall_score(y_test_enc, y_pred, average='macro'):.2f}"

        if hasattr(model, "predict_proba"):
            y_proba = model.predict_proba(X_test_enc)
            results[model_name]["Log Loss"] = f"{log_loss(y_test_enc, y_proba):.3f}"
        else:
            results[model_name]["Log Loss"] = "N/A"

        print(f"\n{model_name} - Confusion Matrix:")
        print(confusion_matrix(y_test_enc, y_pred))

        print(f"\n{model_name} - Classification Report:")
        print(classification_report(y_test_enc, y_pred,zero_division=0, digits=3))

        print(f"{model_name} completed training\n")

    # Summary printout
    model_list = list(results.keys())
    for name in model_list:
        print(f"{name}")
        for metric, val in results[name].items():
            print(f"{metric} - {val}")
        print("\n")

    # Sort models by metrics
    accuracy_list = sorted(model_list, key=lambda x: float(results[x]["Accuracy"].strip('%')), reverse=True)
    f1_list = sorted(model_list, key=lambda x: float(results[x]["F1 (macro)"]), reverse=True)
    f2_list = sorted(model_list, key=lambda x: float(results[x]["F2 Score"]), reverse=True)

    print(f"From highest accuracy to lowest: {accuracy_list}")
    print(f"From highest F1 (macro) score to lowest: {f1_list}")
    print(f"From highest F2 score to lowest: {f2_list}")
```

```python
# run models through dataframes
dframes = [df_data, df_drop_MI, df_drop_RH]
for dataframe in dframes:
    run_models(dataframe)
```
